# Core Models
> Base dataclasses for session entries

```
#| default_exp core
```

```
#| export
from dataclasses import dataclass, field
from datetime import datetime, timezone
from typing import Optional, Union, List, Dict, Any
```

```
#| export
def now_iso() -> str:
    return datetime.now(timezone.utc).isoformat()

def gen_id() -> str:
    import uuid
    return uuid.uuid4().hex[:8]
```

```
#| export
@dataclass
class SessionHeader:
    type: str = 'session'
    version: int = 1
    id: str = field(default_factory=gen_id)
    timestamp: str = field(default_factory=now_iso)
    cwd: str = ''
    parentSession: Optional[str] = None
```

```
#| export
@dataclass
class MessageEntry:
    type: str = 'message'
    id: str = field(default_factory=gen_id)
    parentId: Optional[str] = None
    timestamp: str = field(default_factory=now_iso)
    role: str = ''
    content: Union[str, List[Dict[str, Any]]] = ''
```

```
#| export
@dataclass
class CompactionEntry:
    type: str = 'compaction'
    id: str = field(default_factory=gen_id)
    parentId: Optional[str] = None
    timestamp: str = field(default_factory=now_iso)
    summary: str = ''
    firstKeptEntryId: str = ''
```

```
#| export
@dataclass
class ModelChangeEntry:
    type: str = 'model_change'
    id: str = field(default_factory=gen_id)
    parentId: Optional[str] = None
    timestamp: str = field(default_factory=now_iso)
    provider: str = ''
    modelId: str = ''
```

```
#| export
@dataclass
class ThinkingLevelChangeEntry:
    type: str = 'thinking_level_change'
    id: str = field(default_factory=gen_id)
    parentId: Optional[str] = None
    timestamp: str = field(default_factory=now_iso)
    thinkingLevel: str = 'off'
```

```
#| export
@dataclass
class CustomEntry:
    type: str = 'custom'
    id: str = field(default_factory=gen_id)
    parentId: Optional[str] = None
    timestamp: str = field(default_factory=now_iso)
    customType: str = ''
    data: Any = None
```

```
#| export
@dataclass
class LabelEntry:
    type: str = 'label'
    id: str = field(default_factory=gen_id)
    parentId: Optional[str] = None
    timestamp: str = field(default_factory=now_iso)
    targetId: str = ''
    label: Optional[str] = None
```
